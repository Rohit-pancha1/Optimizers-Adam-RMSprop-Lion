```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop('bought_insurance', axis='columns'), df.bought_insurance)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
x_train['age'] = x_train['age']/100
x_test['age'] = x_test['age']/100
```

```python
x_train.head()
```

```python
from tensorflow import keras
```

```python
model = keras.Sequential(
    [keras.layers.Dense(1, input_shape=(2,), activation= 'sigmoid', bias_initializer = 'zeros')]
)
```

```python
model.compile(optimizer= 'adam', loss ='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs =1000)
```

```python
model.evaluate(x_test, y_test)
```

```python
model.predict(x_test)
```

```python
import numpy as np
def sigmoid(x):
    return(1/(1+np.exp(-x)))
```

```python
def pred_f(age, affordability):
    summa = coef[0]*age + coef[1]*affordability + intercept
    return(sigmoid(summa))
```
